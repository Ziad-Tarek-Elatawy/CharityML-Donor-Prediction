```python
import numpy as np

# Tanh activation function
def tanh(x):
    return np.tanh(x)

# Define input values
i1, i2 = 0.05, 0.10  # Given input values

# Randomly initialize weights from [-0.5, 0.5]
np.random.seed(42)  # For reproducibility
w1, w2, w3, w4 = np.random.uniform(-0.5, 0.5, 4)  # Weights for hidden layer
w5, w6, w7, w8 = np.random.uniform(-0.5, 0.5, 4)  # Weights for output layer

# Biases
b1, b2 = 0.5, 0.7  

# Forward pass for hidden layer
net_h1 = w1 * i1 + w2 * i2 + b1
net_h2 = w3 * i1 + w4 * i2 + b1
out_h1 = tanh(net_h1)
out_h2 = tanh(net_h2)

# Forward pass for output layer
net_o1 = w5 * out_h1 + w6 * out_h2 + b2
net_o2 = w7 * out_h1 + w8 * out_h2 + b2
out_o1 = tanh(net_o1)
out_o2 = tanh(net_o2)

# Print results
print(f"Hidden Layer Outputs: h1 = {out_h1:.6f}, h2 = {out_h2:.6f}")
print(f"Output Layer Outputs: o1 = {out_o1:.6f}, o2 = {out_o2:.6f}")
```
